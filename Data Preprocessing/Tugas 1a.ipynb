# Cek Data Kosong / Mengatasi data null

```python
data_missing_value = df.isnull().sum().reset_index()
data_missing_value.columns = ['feature','missing_value']
data_missing_value['percentage'] = round((data_missing_value['missing_value']/len(df))*100,3)
data_missing_value = data_missing_value.sort_values('percentage', ascending=False).reset_index(drop=True)
data_missing_value = data_missing_value[data_missing_value['percentage']>0]
data_missing_value
```

# Membuang data yang tidak diperlukan

```python
df.columns
```

```python
selected_columns= ['hotel', 'is_canceled', 'lead_time', 'arrival_date_year',
       'arrival_date_month', 'arrival_date_week_number',
       'arrival_date_day_of_month', 'stays_in_weekend_nights',
       'stays_in_weekdays_nights', 'adults', 'children', 'meal',
       'city', 'market_segment', 'distribution_channel',
       'is_repeated_guest', 'previous_cancellations',
       'previous_bookings_not_canceled', 'deposit_type', 'customer_type', 'adr',
       'required_car_parking_spaces', 'total_of_special_requests']
```

#### <font color='blue'> Membuat copy dataframe agar tidak mengganggu dataset asli serta menghapus beberapa data yang tidak di anggap penting  </font>
    1. *Agent* karena Range Max dan Min Yang sangat jauh bisa dianggap nilainya tidak normal dan tidak jelas
    2. *babies* karna di anggap terhitung juga di children, 
    3. *previous_cancellations* Riwayat pembatalan Cancelation pada suatu info,
    4. *days_in_waiting_list* tidak akan memberikan informasi 
    5. *reservation_status* tidak memberikan informasi tambahan 

```python
bookings_df = df[selected_columns].copy()
```

####  <font color='blue'> Feature Extraction : Menjumlahkan adults + children menjadi fitur baru total pengunjung dan lalu menghapus nya ( adults, children ) </font>

```python
bookings_df['Total_Guests'] = bookings_df['adults'] + bookings_df['children']
bookings_df = bookings_df.drop(['adults', 'children'], axis=1)
```

#### <font color='blue'> Melakukan filter data yang hanya memiliki jumlah *total_guest* lebih besar atau sama dengan 1, Karena masih terdapat data tamu yang bernilai 0 tetapi status *is_canceled* masih bernilai 0 </font>

```python
tamu_kosong = bookings_df[bookings_df['Total_Guests'] == 0]
tamu_kosong.head()
```

```python
bookings_df= bookings_df[bookings_df['Total_Guests'] >= 1]
```

```python
bookings_df.head(10)
```

#### melakukan filter data yang tidak ada pembatalan sewa ( is_canceled = 0 ) 

```python
df_no_cancel= bookings_df[bookings_df['is_canceled'] == 0]
df_no_cancel.sample(5)
```

#### <font color='blue'> Memisahkan dan membuat kumpulan data yang berbeda dari jenis hotel, lalu menyimpannya ke file csv ke untuk penelitian lanjutan </font>

```python
df_no_cancel['hotel'].unique()
```

```python
resort_df = df_no_cancel[df_no_cancel['hotel'] == "Resort Hotel"]
city_df= df_no_cancel[df_no_cancel['hotel'] == "City Hotel"]
```

```python
resort_df.to_csv('resort_df.csv',header=True)
city_df.to_csv('city_df.csv',header=True)
```
